### Exercício

```python
ladoA=int(input("Tamanho lado A: "))
ladoB=int(input("Tamanho lado B: "))
ladoC=int(input("Tamanho lado C: "))

if ((ladoA>0 and ladoB>0 and ladoC>0) and (ladoA + ladoB > ladoC and ladoA + ladoC > ladoB and ladoB + ladoC > ladoA)):
  if (ladoA == ladoB) and (ladoB == ladoC):
    print('Triângulo equilátero!')
  elif (ladoA == ladoB) or (ladoB == ladoC) or (ladoC == ladoA):
    print('Triângulo isósceles!')
  else:
    print('Triângulo escaleno!')
else:
  print('Não forma um triângulo!')
```
